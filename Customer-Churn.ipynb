```python
import matplotlib.pyplot as plt
import seaborn as sns

n_col = 3
n_rows = 5

fig, axes = plt.subplots(n_rows, n_col, figsize=(16, 16))
axes = axes.flatten()

for i in range(len(columns)):
    ax = sns.countplot(data=df, x=columns[i], hue='Churn', ax=axes[i])
    axes[i].set_title(f'Count Plot of {columns[i]}')

    # Get total count per group for percentage calculation
    total_per_category = df[columns[i]].value_counts()

    for container in ax.containers:
        # Compute total count for each bar group (hue value)
        labels = []
        for bar in container:
            height = bar.get_height()
            x_val = bar.get_x() + bar.get_width() / 2
            x_category = round(x_val)  # approximate index for x-axis category
            
            # Get category name from bar's x position
            category_idx = int(bar.get_x() + bar.get_width() / 2)
            category_name = bar.get_x()
            
            # Get corresponding x-tick label (safer way)
            xtick_labels = [tick.get_text() for tick in ax.get_xticklabels()]
            if xtick_labels:
                category_name = xtick_labels[container.index(bar) % len(xtick_labels)]

            # Compute percentage
            total = df[df[columns[i]] == category_name].shape[0]
            percent = (height / total) * 100 if total > 0 else 0
            labels.append(f'{percent:.1f}%')

        ax.bar_label(container, labels=labels)

plt.tight_layout()
```

### 🔹 1. SeniorCitizen

* Observation: Senior citizens show a higher churn rate compared to non-seniors.

* Insight: Older customers may face challenges with digital processes or experience-related issues, highlighting the need for personalized assistance and simplified services.

### 🔹 2. Partner

* Observation: Customers with partners churn significantly less.

* Insight: Married or partnered customers tend to have more stability, leading to consistent usage and longer retention.

### 🔹 3. Dependents

* Observation: Customers with dependents have a much lower churn rate.

* Insight: Family-oriented users are more likely to value reliable services and avoid switching providers.

### 🔹 4. PhoneService

* Observation: Churn rates are nearly identical for customers with and without phone services.

* Insight: Traditional phone services are no longer a decisive retention factor; customers focus more on internet-based offerings.

### 🔹 5. MultipleLines

* Observation: Having multiple lines slightly reduces churn, but the effect is small.

* Insight: Bundling helps, but additional service value or discounts may be needed to make a stronger impact.

### 🔹 6. InternetService

* Observation: Fiber optic users show the highest churn, while DSL users have the lowest.

* Insight: Higher pricing or service expectations for fiber optic connections could be driving dissatisfaction.

### 🔹 7. OnlineSecurity

* Observation: Customers with online security have the lowest churn rates.

* Insight: Security and protection features create a sense of trust and safety, which strongly encourages loyalty.

### 🔹 8. OnlineBackup

* Observation: Online backup services correlate with lower churn.

* Insight: Offering convenient data protection and recovery options enhances customer confidence in the service provider.

### 🔹 9. DeviceProtection

* Observation: Customers with device protection are less likely to churn.

* Insight: Providing device-related insurance or protection builds peace of mind and satisfaction.

### 🔹 10. TechSupport

* Observation: Tech support has one of the strongest impacts on retention. Customers without tech support churn significantly more.

* Insight: Quick and reliable support is a critical retention driver — customer service quality directly affects loyalty.

### 🔹 11. StreamingTV

* Observation: Customers with streaming TV churn slightly more.

* Insight: Entertainment add-ons alone do not guarantee retention; service quality and value still matter more.

### 🔹 12. StreamingMovies

* Observation: Customers with streaming movie services show slightly lower churn.

* Insight: Bundled entertainment options add marginal retention benefits when combined with core services.

### 🔹 13. Contract

* Observation: Contract type is the single most influential feature:

  i) Month-to-month contracts → highest churn (~43%)

  ii) One-year → moderate (~11%)

  iii) Two-year → lowest (~3%)

* Insight: Longer-term contracts strongly encourage commitment and stability.

### 🔹 14. PaperlessBilling

* Observation: Paperless billing users churn more than those with paper bills.

* Insight: Although convenient, digital billing may reduce engagement or the “personal touch” some customers value.

### 🔹 15. PaymentMethod

* Observation: Payment method strongly affects churn:

  i) Highest churn: Electronic check

  ii) Lowest churn: Automatic bank transfer / credit card

* Insight: Ease, automation, and reliability in payments enhance retention, while manual or inconvenient payment methods correlate with churn.

# Conclusion

### * Key Findings:
i) Customers are more loyal when they feel supported (tech support, online security, device protection).

ii) Long-term contracts dramatically reduce churn.

iii) Simpler, automated payments and value-added services improve satisfaction.

iv) Demographic and lifestyle factors (like age, dependents, and partners) also influence loyalty.

### * Business Implications:
i) Focus on Retention Programs for senior citizens and month-to-month users.

ii) Encourage service bundling (security, backup, protection) through discounts.

iii) Promote longer contracts with loyalty rewards or flexible renewal options.

iv) Improve customer support and automate billing for frictionless experiences.
